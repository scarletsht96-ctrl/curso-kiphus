1. Create a Pandas DataFrame with the following data:

| Name    | Age | Department |
|---------|-----|------------|
| Alice   | 24  | HR         |
| Bob     | 30  | IT         |
| Charlie | 22  | Marketing  |

Display the DataFrame.

Expected output:

    Name	Age	Department

0	Alice	24	HR

1	Bob	30	IT

```python

import pandas as pd

import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [24, 30, 22],
    "Department": ["HR", "IT", "Marketing"]
}

df = pd.DataFrame(data)
df


```
